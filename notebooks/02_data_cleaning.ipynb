```python
import pandas as pd

rent_df = pd.read_csv("../data/processed/private_rents_clean_initial.csv")

rent_df.head()
```

```python
rent_df.shape
```

```python
rent_df.columns = (
    rent_df.columns
    .str.lower()
    .str.strip()
    .str.replace(" ", "_")
    
)
rent_df.columns
```

```python
rent_clean = rent_df[
    [
        "time_period",
        "area_code",
        "area_name",
        "region_or_country_name",
        "index",
        "monthly_change",
        "annual_change",
        "rental_price"
    ]
].copy()

rent_clean.head()
```

```python
rent_clean.info()
```

```python
rent_clean["time_period"] = pd.to_datetime(rent_clean["time_period"])
```

```python
rent_clean.info()
```

```python
rent_clean = rent_clean.replace(["[x]", "[z]", "[c]"], pd.NA)
```

```python
numeric_cols = [
    "index",
    "monthly_change",
    "annual_change",
    "rental_price"
]
```

```python
for col in numeric_cols:
    rent_clean[col] = pd.to_numeric(rent_clean[col], errors='coerce')
```

```python
rent_clean.info()
```

```python
rent_clean.isna().sum()
```

```python
rent_clean = rent_clean.dropna(subset=["rental_price"])
```

```python
essential_cols = [
    "time_period",
    "area_code",
    "area_name",
    "rental_price"
]

rent_clean = rent_clean.dropna(subset=essential_cols)
```

```python
rent_clean = rent_clean.dropna(
    subset=[
        "time_period",
        "area_code",
        "area_name",
        "rental_price"
    ]
)
```

```python
rent_clean.isna().sum()
```

```python
rent_clean.to_csv(
    "../data/processed/private_rents_clean.csv",
    index=False
)
```
